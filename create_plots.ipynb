```python
import json
import matplotlib.pyplot as plt
```

```python
with open('/home/karakanaidu/experiment/grace/GRACE/results_scotus_grace_eps1.0.json') as f:
    ls = json.load(f)
```

```python
ls[0].keys()
```

```python
nkeys_ls = [x['nkeys'] for x in ls]
TRR_ls = [x['TRR'] for x in ls]
ERR_ls = [x['ERR'] for x in ls]
n_edits = [x['n_edits'] for x in ls]
```

```python
# Plot 1: nkeys_ls vs n_edits
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.plot(n_edits, nkeys_ls, color='b')
plt.title('Number of Keys vs Number of Edits')
plt.xlabel('Number of Edits')
plt.ylabel('Number of Keys')

# Plot 2: TRR_ls vs n_edits
plt.subplot(1, 3, 2)
plt.plot(n_edits, TRR_ls, color='g')
plt.title('TRR vs Number of Edits')
plt.xlabel('Number of Edits')
plt.ylabel('TRR')

# Plot 3: ERR_ls vs n_edits
plt.subplot(1, 3, 3)
plt.plot(n_edits, ERR_ls, color='r')
plt.title('ERR vs Number of Edits')
plt.xlabel('Number of Edits')
plt.ylabel('ERR')

plt.tight_layout()
plt.show()
```
